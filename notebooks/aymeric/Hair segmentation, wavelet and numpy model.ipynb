```python
def mean(chan_list, coef):
    return np.expand_dims(np.mean(np.stack([chan_list[i][coef] for i in range(3)], axis = -1), axis = 2),axis = -1)
```

```python
def apply_filter(kind, image):
    chan_3 = []
    for i in range(3):
        coeffs2 = pywt.dwt2(image[:,:,i], kind)
        ll, (lh, hl, hh) = coeffs2
        chan_3.append((ll, lh, hl, hh))
    LL = 255-mean(chan_3, 0)
    LH = 255-mean(chan_3, 1)
    HL = 255-mean(chan_3, 2)
    HH = 255-mean(chan_3, 3)
    return LL, LH, HL, HH
```

```python
def apply_filter_on_channel(kind, image):

    coeffs2 = pywt.dwt2(image, kind)
    LL, (LH, HL, HH) = coeffs2
    return LL, LH, HL, HH
```

```python
def print_waves(kind, image):
    # Wavelet transform of image, and plot approximation and details
    titles = ['Approximation', ' Horizontal detail',
              'Vertical detail', 'Diagonal detail']
    LL, LH, HL, HH = apply_filter(kind, image)
    fig = plt.figure(figsize=(12, 3))
    for i, a in enumerate([LL, LH, HL, HH]):
        ax = fig.add_subplot(1, 4, i + 1)
        ax.imshow(a.astype("uint8"), interpolation="nearest", cmap=plt.cm.gray)
        ax.set_title(titles[i], fontsize=10)
        ax.set_xticks([])
        ax.set_yticks([])
    fig.suptitle(kind, fontsize=16)
    fig.tight_layout()
    plt.show()
```

```python
def print_waves_double(kind, image):
    # Wavelet transform of image, and plot approximation and details
    titles = ['Approximation', ' Horizontal detail',
              'Vertical detail', 'Diagonal detail']
    LL, LH, HL, HH = apply_filter(kind, image)
    LL, LH, HL, HH = apply_filter_on_channel(kind, LL)
    fig = plt.figure(figsize=(12, 3))
    for i, a in enumerate([LL, LH, HL, HH]):
        ax = fig.add_subplot(1, 4, i + 1)
        ax.imshow(a.astype("uint8"), interpolation="nearest", cmap=plt.cm.gray)
        ax.set_title(titles[i], fontsize=10)
        ax.set_xticks([])
        ax.set_yticks([])
    fig.suptitle(kind, fontsize=16)
    fig.tight_layout()
    plt.show()
```

```python
def rotate_numpy_image(img):
    new_img = np.empty(img.shape)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            new_img[j,img.shape[1]-1-i] = img[i,j]
    return new_img
```

```python
def apply_filter_grey(kind, image):
    ll, (lh, hl, hh)  = pywt.dwt2(image, kind)
    LL = 255-ll
    LH = 255-lh
    HL = 255-hl
    HH = 255-hh
    return LL, LH, HL, HH
```

```python
def print_waves_grey(kind, image):
    # Wavelet transform of image, and plot approximation and details
    titles = ['Approximation', ' Horizontal detail',
              'Vertical detail', 'Diagonal detail']
    LL, LH, HL, HH = apply_filter_grey(kind, image)
    fig = plt.figure(figsize=(12, 3))
    for i, a in enumerate([LL, LH, HL, HH]):
        ax = fig.add_subplot(1, 4, i + 1)
        ax.imshow(a.astype("uint8"), interpolation="nearest", cmap=plt.cm.gray)
        ax.set_title(titles[i], fontsize=10)
        ax.set_xticks([])
        ax.set_yticks([])
    fig.suptitle(kind, fontsize=16)
    fig.tight_layout()
    plt.show()
```

```python
def std_of_col_means(img):
    edges = prewitt(img).astype("uint8")
    edges[edges<123] = 0
    edges[edges>122] = 255
    list_of_col_means = []
    mask = (filter(edges,3) !=0)
    for i in range(edges.shape[1]):
        hair_col = edges[:,i][mask[:,i]]
        if hair_col.any():
            list_of_col_means.append(hair_col.mean())
    return np.std(list_of_col_means)
```
